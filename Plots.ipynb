```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Code and data used to generate Figure 4.3
plt.figure(figsize = (7.5,5))
x = np.linspace(0,0.25,28265)
npzfile = np.load('data_unde.npz')
npzfile.files
plt.plot(x, npzfile['unde'])
npzfile = np.load('data_crit.npz')
npzfile.files
plt.plot(x, npzfile['crit'])
npzfile = np.load('data_over.npz')
npzfile.files
plt.plot(x, npzfile['over'])
plt.xlabel('adhesive/contact thermal diffusivity')
plt.ylabel('Critical Power (arb. units)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature ($^\circ$C)')
plt.legend(('over-powered', 'critically-powered', 'under-powered'), prop={'size': 11})
#plt.savefig('Figure4_3.png', dpi=300)
```
